## 파이썬 클래스
- 파이썬 클래스는 파이썬에서 타입을 만들 때 사용하는 키워드입니다.
파이썬의 기본 데이터 타입인 int, float, str과 데이터 저장에 사용되는
리스트, 튜플, 딕셔너리 등이 모두 클래스로 만들어진 타입입니다

```python
a =3
b = "hello"

print(type(a))
print(type(b))
```

- 파이썬 초보자가 클래스를 사용해서 직접 타입을 만들어야 하는 경우는 많지 않습니다.
- 다만 파이썬의 여러 라이브러리에서 각자의 데이터를 효과적으로 다루기 위해 
타입을 만들어서 사용하기 때문에 우리가 클래스를 배우지 않고 서는 
해당 라이브러리를 문법적으로 정확히 이해할 수 없습니다.

- 예를 들어 어떤 변수가 문자열 객체를 바인딩하고 있을 떄 우리는 점(,)을 찍고 upeer().lower()같은 메서드를 호출하여 문자열을 조작할 수 있었습니다.

- 이 모든 것들이 이미 문자열에서 자주 사용되는 기능들을 메서드로 만들고
이를 문자열(str)클래스로 정의했기 때문입니다

- 일반적으로 프로그램이라는 것은 데이터와 데이터를 처리하는 함수로 구성됩니다.
- 파이썬 클래스는 데이터와 데이터를 처리하는 함수를 하나의 공간(클래스)에 정의한 것입니다.
- 클래스 안에 정의된 함수를 특별히 메서드라고 부릅니다.

- 정리해보면 클래스 = 데이터 + 함수로 구성되며 '새로운 타입'을 만드는데 사용합니다

#### 붕어빵 클래스
- 붕어빵을 구울 때 붕어빵을 손으로 만들어서 직접 굽지 않고 붕어빵 틀을 사용합니다.
- 이와 비슷하게 프로그래밍을 하다 보면 문자열을 다룰 일이 자주 있는데 이 때 문에 기본적으로 문자열 클래스를 제공하는 겁니다.
- 붕어빵 틀 클래스를 만들어봅시다
- 클래스 안에는 함수가 있을 수 있는데 일단 기능이 생각나지 않으므로 pass라고 적어주면 문법 에러는 나지 않습니다.
- 클래스를 정의하면 클래스 이름이 '클래스 객체'를 바인딩 합니다


```python
class 붕어빵틀:
    pass
```

- 붕어빵틀이라는 클래스에는 어떤 기능(함수)이 있어야 할까요?
- '앙꼬넣기' 함수를 추가해보겠습니다
- 클래스 안에 정의된 함수를 메서드라고 부르고 첫번째 인자는 관례적으로 self를 사용하겠습니다.
- '앙꼬넣기' 함수는 함수 인자로 부터 '앙꼬'를 받아서 이를 붕어빵에 넣어줍니다
- 붕어빵틀 클래스에 메서드를 하나 추가했습니다. 메모리의 상태는 이렇습니다


```python
class 붕어빵틀:
    def 앙꼬넣기(self, 앙꼬):
        self.앙꼬 = 앙꼬
```

- 붕어빵 틀을 만들었다면 붕어빵 틀로부터 붕어빵을 구울 수 있습니다.
- 프로그래밍에서도 클래스를 정의했다면 클래스로부터 인스턴스를 새애성할 수 있습니다.
- 붕어빵 틀이 붕어빵 클래스의 인스턴스입니다.


```python
class 붕어빵틀:
    def 앙꼬넣기(self, 앙꼬):
        self.앙꼬 = 앙꼬
붕어빵1 = 붕어빵틀() #붕어빵1,2라는 변수에 바인딩된 붕어빵틀 타입의 생성된 객체
붕어빵2 = 붕어빵틀()
```

- 파이썬 인터프리터가 위 코드를 실행하면 클래스 객체1개와 인스턴스 2개를 메모리에 생성해줍니다.
- 클래스 객체 내에는 앙꼬넣기라는 메서드가 존재하고 인스턴스 2개에는 아무 값도 존재하지 않습니다.
- 클래스 객체로부터 인스턴스 생성


- 붕어빵을 구웠는데 앙꼬가 들어있지 않습니다.
- 붕어빵틀 클래스에 앙꼬넣기라는 메서드를 사용해서 넣어보겠습니다

```python
class 붕어빵틀:
    def 앙꼬넣기(self,앙꼬):
        self.앙꼬 = 앙꼬
        
붕어빵1 = 붕어빵틀() # 
붕어빵틀.앙꼬넣기(붕어빵1, "초코맛") #  붕어빵틀 클래스 안의 (.의 역할) 앙꼬넣기 메서드를 호출합니다.
# self 인자에 붕어빵1이 가리키는 인스턴스를 앙꼬 인자에 “초코맛”이라는 문자열을 전달합니다. 

붕어빵2 = 붕어빵틀()
붕어빵틀.앙꼬넣기(붕어빵2, "딸기맛")
```

- 클래스 안에 정의된 앙꼬넣기 메서드를 호출하면 메서드 호출시 전달 된 앙꼬가 인스턴스에 저장됩니다


메서드 호출에 대한 코드가 잘 이해되지 않는다면 앞서 배운것처럼 메서드가 호출되는 단계를 그림으로 그려보면 됩니다. 3단계로 나눠서 이해해 봅시다

- 1단계 : 메서드를 호출하면 코드의 실행 흐름이 앙꼬넣기() 메서드로 이동합니다.
- 메서드에는 2개의 인자가 있습니다.
- self 변수는 붕어빵1이 바인딩하는 값을 바인딩 합니다.
- 앙꼬라는 변수는 호출부로부터 전달된 문자열인 '초코맛'을 바인딩합니다.
- 메모리의 상태는 

- 함수(메서드) 호출 2단계는 '코드 실행'입니다.
- 앙꼬넣기 함수의 실행 코드는 'self.앙꼬 = 앙꼬'입니다.
-  ‘=’이 있으므로 오른쪽을 먼저 해석합니다. 
- 현재 앙꼬라는 변수는 ‘초코맛’이라는 문자열을 바인딩하고 있는 상태입니다.
이를 self.앙꼬가 바인딩하면 됩니다. 
- 현재 self는 ‘붕어빵1’이라는 객체를 바인딩하고 있습니다. 따라서 해당 객체에 ‘앙꼬’라는 변수가 생기고 이 변수가 “초코맛”을 바인딩합니다.
- 파이썬에서는 객체에 데이터가 딕셔너리로 저장됩니다.
- 딕셔너리의 key에 변수 이름이 딕셔너리의 value에 값이 저장됩니다.
